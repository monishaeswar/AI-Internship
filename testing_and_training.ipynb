```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([5, 6, 1, 0, 3])

model = LinearRegression()
model.fit(X, y)

new_input_data = np.array([[6]])
new_prediction = model.predict(new_input_data)

plt.scatter(X, y, color='blue', label='Training Data')
plt.scatter(new_input_data, new_prediction, color='red', label='Prediction (x=6)')

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Prediction")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([7, 8, 1, 0, 2])

model = LinearRegression()
model.fit(X, y)

new_input_data = np.array([[6]])
new_prediction = model.predict(new_input_data)

plt.scatter(X, y, color='blue', label='Training Data')
plt.scatter(new_input_data, new_prediction, color='red', label='Prediction (x=6)')

plt.xlabel("X")
plt.ylabel("y")
plt.bar(X.flatten(),y)
plt.bar(new_input_data.flatten(), new_prediction.flatten(), color='red', label='Prediction (x=6)')
plt.title("Linear Regression Prediction")
plt.plot(X,y)
plt.legend()
plt.grid(True)
plt.show()
print("Predictions:", new_prediction)
print("new_Predictions:", new_prediction)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([[1], [2], [3], [4], [5]])
y = np.array([7, 8, 1, 6, 2])

model = LinearRegression()
model.fit(X, y)

# Fix: Reshape new_input_data to be a 2D array
new_input_data = np.array([6]).reshape(-1, 1)
new_prediction = model.predict(new_input_data)

plt.scatter(X, y, color='blue', label='Training Data')
plt.scatter(new_input_data, new_prediction, color='red', label='Prediction (x=6)')

plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression Prediction")
plt.plot(X,y)
plt.legend()
plt.grid(True)
plt.show()
print("Predictions:", new_prediction)
print("new_Predictions:", new_prediction)
```
