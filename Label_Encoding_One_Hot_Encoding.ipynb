```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.DataFrame({
    "Name": ["Ram", "Priya", "Arun", "Kavi"],
    "Gender": ["Male", "Female", "Male", "Female"]
})

print(df)
```

```python
le = LabelEncoder()

df["Gender"] = le.fit_transform(df["Gender"])

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Ram", "Priya", "Arun", "Kavi"],
    "Gender": ["Male", "Female", "Male", "Female"]
})

print(df)
```

```python
encoded = pd.get_dummies(df, columns=["Gender"], dtype=int)

print(encoded)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Gender": ["Male", "Female", "Male"],
    "City": ["Chennai", "Vellore", "Bangalore"]
})

encoded = pd.get_dummies(df, dtype=int)

print(encoded)
```

```python
df = pd.DataFrame({
    "Color": ["Red", "Blue", "Green", "Blue", "Red"],
    "Size": ["Small", "Medium", "Large", "Small", "Large"]
})
encoded = pd.get_dummies(df, dtype=int)

print(encoded)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Student": ["A", "B", "C", "D", "E"],
    "Department": ["CSE", "ECE", "IT", "CSE", "EEE"]
})

print(df)
```

```python
from sklearn.preprocessing import LabelEncoder

# Create a copy of the dataset
label_df = df.copy()

le = LabelEncoder()

label_df["Department"] = le.fit_transform(label_df["Department"])

print(label_df)
```

```python
print(le.classes_)
```

```python
onehot_df = pd.get_dummies(df, columns=["Department"], dtype=int)
print(onehot_df)
```
